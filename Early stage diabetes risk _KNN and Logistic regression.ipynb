```python
#importing required library
import pandas as pd
import numpy as np
```

```python
#reading train and test data 
training_data= pd.read_excel(r'C:\Users\harita\Mypython\Machine learning\KNN\Early diabetes_train.xlsx',header=0)
testing_data= pd.read_excel(r'C:\Users\harita\Mypython\Machine learning\KNN\Early diabetes_test.xlsx',header=0)

```

```python
#loading train data head to check it been imported properly
training_data.head()
```

```python
#loading test data head to check it been imported properly
testing_data.head()
```

```python
#checking shape and basic info of the data
print(training_data.shape)
print(testing_data.shape)

print(training_data.info())
print(testing_data.info())
```

```python
#Checking wheather the train and test data has Missing Values

print(training_data.isnull().sum())
print(testing_data.isnull().sum())
```

```python
#"Age", "sudden weight loss" and "muscle stiffness" has missg values in both train and test data. 
# Replacing "sudden weight loss"and "muscle stiffness" variables with mode value because they are categorical.
# Replacing "Age" variable with mean value because they are numeric

training_data["Age"].fillna(int(training_data["Age"].mean()),inplace=True)
training_data["sudden weight loss"].fillna(training_data["sudden weight loss"].mode()[0],inplace=True)
training_data["muscle stiffness"].fillna(training_data["muscle stiffness"].mode()[0],inplace=True)
                                         
testing_data["Age"].fillna(int(testing_data["Age"].mean()),inplace=True)
testing_data["sudden weight loss"].fillna(testing_data["sudden weight loss"].mode()[0],inplace=True)
testing_data["muscle stiffness"].fillna(testing_data["muscle stiffness"].mode()[0],inplace=True)

print(training_data.isnull().sum())
print(testing_data.isnull().sum())
```

```python
#We are no checking for the Outliers because KNN doesnot affected by outlier. never going to affect our model
```

```python
#Converting categorical variable into numerical using lable encoder

colname=[]
for x in training_data.columns:
    if training_data[x].dtype=='object':
        colname.append(x)
colname
```

```python
from sklearn import preprocessing

le=preprocessing.LabelEncoder()

for x in colname:
    training_data[x]=le.fit_transform(training_data[x])
    testing_data[x]=le.fit_transform(testing_data[x])
```

```python
#checking wheather above steps worked properly
training_data.head()
```

```python
testing_data.head()
```

```python
#creating X and Y variables

X_train = training_data.values[:,0:-1]
X_test = testing_data.values[:,0:-1]

Y_train = training_data.values[:,-1]
Y_test = testing_data.values[:,-1]

Y_train= Y_train.astype(int)
Y_test= Y_test.astype(int)
```

```python
print(X_train)
print(Y_train)
```

```python
#scaling is necessary to get data in one range

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler.fit(X_train)
# while working on seperate train and test data fit scaler object only upon the train data
#scaler object will learn for train data
#and use the same scaler object to transfer both train as well as test data


X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

## KNN

```python
#now we are all set to run the model
# Running KNN (sqrt approach) model 

import numpy as np

from sklearn.neighbors import KNeighborsClassifier

model_KNN= KNeighborsClassifier(n_neighbors= int(np.sqrt(len(X_train))),
                                    metric= 'euclidean')

#fit the model on the data and predict the values

model_KNN.fit(X_train,Y_train)

Y_pred= model_KNN.predict(X_test)

print(list(zip(Y_test,Y_pred)))
```

```python
#Model Evaluation

from sklearn.metrics import confusion_matrix, accuracy_score,classification_report

cfm=confusion_matrix(Y_test,Y_pred)
print(cfm)
print()

print("Classification report: ")

print(classification_report(Y_test,Y_pred))

acc=accuracy_score(Y_test, Y_pred)
print("Accuracy of the model: ",acc)
```

```python
# trial n error

from sklearn.metrics import accuracy_score
my_dict={}
for K in range(1,15):
    model_KNN = KNeighborsClassifier(n_neighbors=K,metric="euclidean")
    model_KNN.fit(X_train, Y_train) 
    Y_pred = model_KNN.predict(X_test)
    print ("Accuracy is ", accuracy_score(Y_test,Y_pred), "for K-Value:",K)
    my_dict[K]=accuracy_score(Y_test,Y_pred)
```

## Logistic regression

```python
#Trying out logistic regression
from sklearn.linear_model import LogisticRegression
#create a model
classifier=LogisticRegression()
#fitting training data to the model
classifier.fit(X_train,Y_train)

Y_pred=classifier.predict(X_test)
print(list(zip(Y_test,Y_pred)))
```

```python
#model Evaluation
from sklearn.metrics import confusion_matrix, accuracy_score,classification_report

confusion_matrix=confusion_matrix(Y_test,Y_pred)
print(confusion_matrix)
print()
print("Classification report: ")

print(classification_report(Y_test,Y_pred))

accuracy_score=accuracy_score(Y_test, Y_pred)
print("Accuracy of the model: ",accuracy_score)
```
